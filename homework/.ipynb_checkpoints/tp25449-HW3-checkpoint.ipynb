```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def taylors_sin(n,x):  #approximates sin(x) using the first n terms of taylor's approximation
    value = np.zeros_like(x, dtype = float) #chatgpt says fixing the type of the array fixes weird rounding errors
    for i in range(n):
        value += (-1)**i * x**(2*i+1) / math.factorial(2*i +1)
        
    return value

```

```python
test = np.linspace(0, 2*np.pi, 4)

print(taylors_sin(4, test))
print(np.sin(test))
```

```python
test = np.arange(0,10, 0.1)

#Computes our Sin Values
true = np.sin(test)
one = taylors_sin(1, test)
two = taylors_sin(2, test)
three = taylors_sin(3, test)
four = taylors_sin(4, test)

#Compute Errors
one_error = np.abs(one - true)
two_error = np.abs(two - true)
three_error = np.abs(three -true)
four_error = np.abs(four - true)


#Plots of Sin values
plt.figure()

plt.plot(test, one, label = "1 term")
plt.plot(test, two, label = "2 terms")
plt.plot(test, three, label = "3 terms")
plt.plot(test, four, label = "4 terms")
plt.plot(test, true, label = "True Sin")

plt.title("Taylor Series for sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.ylim(-2, 2)

plt.show()

#PLots of Errors
plt.figure()
plt.plot(test, one_error, label = "1 term")
plt.plot(test, two_error, label = "2 terms")
plt.plot(test, three_error, label = "3 terms")
plt.plot(test, four_error, label = "4 terms")

plt.title("Taylor Series Truncation Error for sin(x)")
plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.legend()
plt.show()

```

```python
plt.figure()
plt.plot(test, one_error, label = "1 term")
plt.plot(test, two_error, label = "2 terms")
plt.plot(test, three_error, label = "3 terms")
plt.plot(test, four_error, label = "4 terms")

plt.title("Taylor Series Truncation Error for sin(x) (Zoomed In) ")
plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.legend()

plt.ylim(0, 20)    # zoom into small errors, asked chatgpt how to zoom in on plots lol

plt.show()
```

The truncation error decreases as more terms are included in the Taylor series approximation. Using only one term gives a poor approximation except very close to $x = 0$. As additional terms are added, the approximation improves significantly for small values of x.

However, notice that the error of higher order approximations blows up significantly for larger values of x, expecially for $x = 10$. For example, the approximation using 4 terms is very accurate up until $x \approx 3.5$, but completely blows up after $x =4$. That is because Taylor series expansions converge best near the point where they are centered, which in this case is $x = 0$. If we want to mitigate the value of the error, we would have to add significantly more terms to the approximation.
