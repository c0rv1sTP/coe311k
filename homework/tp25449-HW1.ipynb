```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def approximate_pi(numSamples):
    counter = 0
    for i in range(numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1, 1)
        
        if((x**2 + y**2 < 1)): 
            counter += 1
    
    return 4 * counter/numSamples
    
```

```python
#iterates sample sizes

x = np.arange(10,10001,20)

true_values = []

for n in x:
    test = approximate_pi(n)
    true_error = abs(np.pi - test)
    
    true_values.append(true_error)
    
    
true_values = np.array(true_values)

plt.plot(x, true_values)
plt.xlabel("Number of Samples")
plt.ylabel("Absolute Error")
plt.show()
```
