### Introduction

The dataset analyzed in this report consists of the U.S. GDP growth rate (in percent) for selected quarters from 2010 to 2023. We focus on three methods to study the data: cubic spline interpolation, polynomial approximation, and least-squares linear regression. The data source is the U.S. Bureau of Economic Analysis (BEA), which provides seasonally adjusted annual rates (SAAR) of real GDP growth.

Below is the vectorized version of the data points:

```python
# Dataset
x = np.array([
    1, 5, 9, 13, 17, 19, 22, 25, 27, 28,
    29, 33, 37, 41, 42, 43, 45, 49, 54, 56
])

y = np.array([
    1.7, 0.1, 2.3, 2.7, 1.7, 5.0, 3.0, 1.5, 3.5, 1.8,
    1.3, 2.5, 3.1, -5.1, -28.1, 33.8, 6.3, -1.6, 2.4, 3.3
])
```

### PART A

### Natural Cubic Spline Construction

First, we use the method of interpolation to analyze our data>

We construct a natural cubic spline by enforcing:

Interpolation: $S_i(x_i) = y_i $

Continuity: S_i(x_{i+1}) = $S_{i+1}(x_{i+1})$

First derivative continuity: $ S_i'(x_{i+1}) = S_{i+1}'(x_{i+1}) $

Second derivative continuity: $S_i''(x_{i+1}) = S_{i+1}''(x_{i+1}) $

Natural boundary conditions:
$S''(x_1) = 0, S''(x_n) = 0$

This leads to a tridiagonal system of size (n−2)×(n−2). Solving this system gives us the necessary constants solve for the full cubic spline.

```python
#Custom Cubic Spline ALgorithm
def setup_tridiagonal_matrix(x, y):
    n = len(x)
    h = np.diff(x)  # Step sizes between data points
    # Initialize the tridiagonal matrix coefficients
    A = np.zeros(n - 3)  # Subdiagonal
    B = np.zeros(n - 2)  # Main diagonal
    C = np.zeros(n - 3)  # Superdiagonal
    D = np.zeros(n - 2)  # Right-hand side vector
    # Populate the tridiagonal matrix coefficients
    
    A = h[1:-1]              # sub-diagonal
    B = 2*(h[:-1] + h[1:]) # main diagonal
    C = h[1:-1]              # super-diagonal
    D = 6 * ((y[2:] - y[1:-1])/h[1:] - (y[1:-1] - y[:-2])/h[:-1]) #RHS vector
    
      
    return A, B, C, D

def thomas_algorithm(a, b, c, d):
    """
    Solve the tridiagonal system using the Thomas algorithm.
    a: sub-diagonal elements (length n-1)
    b: main diagonal elements (length n)
    c: super-diagonal elements (length n-1)
    d: right-hand side (length n)
    Returns: solution vector x of length n
    """
    n = len(b)
    # Forward elimination
    for i in range(1, n):
        w = a[i-1] / b[i-1]
        b[i] = b[i] - w * c[i-1]
        d[i] = d[i] - w * d[i-1]
    # Back substitution
    x = np.zeros(n)
    x[-1] = d[-1] / b[-1]
    
    for i in range(n-2, -1, -1):
        x[i] = (d[i] - c[i] * x[i+1]) / b[i]
    return x

def evaluate_spline(x_data, y_data, x_query):
    
    A, B, C, D = setup_tridiagonal_matrix(x_data,y_data)
    M_interior = thomas_algorithm(A,B,C,D)
    M = np.zeros(len(x)) #curvature of the full data set
    M[1:-1] = M_interior
    
    """M: second derivatives; M[0]=M[-1]=0 for natural spline"""
    n = len(x_data)
    h = np.diff(x_data) # step sizes
    
    i = np.searchsorted(x_data, x_query, side='right') - 1
    i = np.clip(i, 0, n - 2) # find interval
    
    hi = h[i]
    dx_left = x_query - x_data[i] # (x - xᵢ)
    dx_right = x_data[i+1] - x_query # (xᵢ₊₁ - x)
    S = (
        (M[i] / (6*hi)) * dx_right**3 + # Term 1
        (M[i+1] / (6*hi)) * dx_left**3 + # Term 2
        (y_data[i] /hi - M[i] *hi/6) * dx_right + # Term 3
        (y_data[i+1] /hi - M[i+1] *hi/6) * dx_left # Term 4
    )
    return S
```

```python
x_spline = np.linspace(0,60,100)
y_spline = evaluate_spline(x,y, x) # This array is for the purpose of evaluating MSE and R^2
y_spline_graph = evaluate_spline(x,y, x_spline) # This array is for graphing purposes

#Plots data and the spline
plt.plot(x, y, 'bo', label="Data points")
plt.plot(x_spline, y_spline_graph, 'r-',
label="Interpolation")
plt.title("Cubic Spline Interpolation of GDP Growth Dataset")
plt.xlabel("Quarters, Starting from 2010 Q1 (Index 1-56)")
plt.ylabel("GDP Growth %")
plt.legend()
plt.show()
```

### Runge Phenomenon and Outliers

The Runge phenomenon refers to the oscillatory behavior that arises when high-degree polynomials are used to interpolate data exactly, particularly near the boundaries of the interval. This occurs because the polynomial must pass through every data point, causing large oscillations when the data contains extreme values.

In this dataset, the value at 2020 Q2 (−28.1%) represents a significant outlier due to the COVID-19 economic shock. A method that enforces exact interpolation must pass through this point, which can introduce sharp local curvature in the interpolant. While cubic spline interpolation was designed to minimize the Runge phenomenon, the oscillatory behavior is still noticeable near quarters during the COVID pandemic.

This highlights a key trade-off between exact interpolation and numerical stability. While interpolation ensures perfect agreement with observed data, it can lead to unrealistic behavior when the data contains noise or outliers. Approximation methods, such as least squares, provide a smoother fit (as seen below) by minimizing overall error rather than forcing agreement at every point.
In this context, smoothing splines or weighted least squares would be preferable. These methods reduce the influence of extreme values, producing a more stable and realistic representation of the underlying trend while still capturing the general behavior of the data.

### PART B

### Polynomial Least Squares Fit

We fit a degree-4 polynomial using least squares. The model has the form:

$y = β₀ + β₁x + β₂x² + β₃x³ + β₄x⁴$

We construct a Vandermonde matrix A, where each column corresponds to a power of x.

The least squares solution is obtained by solving the normal equations:

$AᵀAβ = Aᵀy$

This method allows us to approximate the data without forcing the model to pass through every point, reducing sensitivity to noise and outliers.

```python
def polynomial_fit(x,y, degree,  x_vals):
    
    n = len(x)
    
    #Build the Vandermonde matrix
    A = np.zeros((n, degree +1))
    
    for i in range(n):
        for j in range(degree + 1):
            A[i][j]= x[i]**j #x^j
            
    # Solve normal Equations
    y_vec = np.array(y)
    
    ATA = np.dot(A.T, A)
    ATy = np.dot(A.T, y_vec)
    
    coeffs = np.linalg.solve(ATA, ATy)
    
    # Calculate predicted values and builds curve
                      
    y_vals = []
    
    for x in x_vals:
        y = 0
        for j in range(len(coeffs)):
            y += coeffs[j] * (x ** j)
        y_vals.append(y)
    
    return np.array(y_vals), A        
```

```python

y_poly, vander = polynomial_fit(x,y,4, x)


# Plot data points and polynomial curve
plt.plot(x, y, 'bo', label="Data points")
plt.plot(x, y_poly, 'r-',
label="Approximation")
plt.title("Polynomial Approximation of GDP Growth Dataset")
plt.xlabel("Quarters, Starting from 2010 Q1 (Index 1-56)")
plt.ylabel("GDP Growth %")
plt.legend()
plt.show()
```

### Interpretation of Data

A degree-4 polynomial fit using least squares provides a smoother approximation of the data compared to the spline fit. When plotted together, the cubic spline better reproduces specific data points, but the polynomial fit is generally more effective at capturing the overall trend. From this result, we are able to see the main advantage of approximation: interpolation prioritizes exact fit and can be sensitive to fluctuations, whereas approximation smooths out noise and is generally more reliable for prediction. However, we'll see later on that this model is not exactly perfect.

### Least Squares Linear Regression

We model the relationship between quarter index \( x \) and GDP growth \( y \) using a linear model:


$
y = \beta_0 + \beta_1 x
$

Since the data does not lie exactly on a line, we compute the best-fit line by minimizing the sum of squared residuals:

$
\min_{\beta_0, \beta_1} \sum_{i=1}^{n} (y_i - (\beta_0 + \beta_1 x_i))^2
$

This leads to the normal equations:

$
A^T A \mathbf{\beta} = A^T \mathbf{y}
$

where

$
A =
\begin{bmatrix}
1 & x_1 \\
1 & x_2 \\
\vdots & \vdots \\
1 & x_n
\end{bmatrix}, \quad
\mathbf{\beta} =
\begin{bmatrix}
\beta_0 \\
\beta_1
\end{bmatrix}
$

We construct $ A $, compute $ A^T A $ and $ A^T y $, and solve the resulting system to obtain the least squares solution.

```python
def least_square(X,Y): #Custom Least square algorithm
    n = len(X)
    
    #Step 1: Summations
    sum_X = np.sum(X)
    sum_Y = np.sum(Y)
    sum_X2 = np.sum(X**2)
    sum_XY = np.sum(X*Y)
    
    #Step 2: Compute m,c
    m = (n * sum_XY - sum_X * sum_Y)/ (n *sum_X2 - sum_X**2)
    c = (sum_Y - m * sum_X) /n
    

    
    return m, c
```

```python
x_no_covid = np.concatenate((x[:13], x[17:]))  #Dataset excluding COVID
y_no_covid = np.concatenate((y[:13], y[17:]))
y_no_covid = np.array(y_no_covid, dtype=float)




slope, intercept = least_square(x_no_covid,y_no_covid)
y_ls = slope*x_no_covid + intercept

# Plot data points and least square curve
plt.plot(x_no_covid, y_no_covid, 'bo', label="Data points")
plt.plot(x_no_covid, y_ls, 'r-',
label="Approximation")
plt.title("Least Square Approximation of GDP Growth Dataset")
plt.xlabel("Quarters, Starting from 2010 Q1 (Index 1-56), Excluding COVID Data")
plt.ylabel("GDP Growth %")
plt.legend()
plt.show()

print("Slope of Regression Line: ", slope)
print("Intercept of Regression Line: ", intercept)
```

### Interpretation of Data

The slope of the least squares linear model is approximately −0.002, indicating a very slight downward trend in GDP growth over time. However, this change is extremely small and not economically significant.
A linear trend does not make strong economic sense for this data. GDP growth is influenced by business cycles, policy changes, and external shocks, leading to fluctuations rather than a steady linear pattern. 
The linear model assumes a constant rate of change over time, independent and random deviations, and no major structural changes in the data. These assumptions are not well satisfied in this context, as GDP growth exhibits cyclical behavior and structural variability.
Therefore, while the linear model provides a simple approximation, it is not an appropriate representation of the economic trend.

```python
# Residual Plot
residual = y_no_covid - y_ls

plt.plot(x_no_covid, residual, 'ro')
plt.title("Residual Plot of the Least Square Approximation")
plt.xlabel("Quarters Index")
plt.ylabel("Residues")
plt.show()

print("Residual Values:", residual)
```

As we can see, the residual values vary wildly. The effects of this variation can be seen in the next section.

### PART C

```python
# Mean Squared Error (MSE)
def mse(y, y_pred):
    mse = np.mean((y - y_pred)**2)
    return mse
# R-squared value
def r_squared(y, y_pred):
    ss_total = np.sum((y - np.mean(y))**2)
    ss_res = np.sum((y - y_pred)**2)
    r_squared = 1 - (ss_res / ss_total)
    return r_squared
```

```python
print("Relevant Error Evaluation Values:")

print("\tCubic Spline:")

print(f'\t\tMean Squared Error: {mse(y, y_spline):.4f}')
print(f'\t\tR-squared: {r_squared(y, y_spline):.4f}')

print("\tPolynomial Fit:")
cond_num = np.linalg.cond(vander, p = None)
print(f"\t\tCondition Number: {cond_num:.2e}")
print(f'\t\tMean Squared Error: {mse(y, y_poly):.4f}')
print(f'\t\tR-squared: {r_squared(y, y_poly):.4f}')

print("\tLeast Square:")
print(f'\t\tMean Squared Error: {mse(y_no_covid, y_ls):.4f}')
print(f'\t\tR-squared: {r_squared(y_no_covid, y_ls):.6f}')
```

### Interpretation of Error Values

The degree-4 polynomial fit results in a nonzero mean squared error, indicating that it does not interpolate the data exactly. This is expected, as least squares fitting minimizes overall error rather than forcing agreement at every point. Due to the exclusion of the data during the COVID pandemic, the least-square linear model yielded a rather reasonable MSE value, indicating that it's a good fit for the data.

The condition number of the Vandermonde matrix was approximately $1.82 * 10^7$, indicating that the system is moderately ill-conditioned. This arises from the rapidly increasing powers of x, which lead to large differences in scale between columns of the matrix. As a result, small numerical errors may be amplified when solving the normal equations, reducing the reliability of the polynomial coefficients. This could also explain the polynomial curve's low $R^2$ value.

The linear least squares model produces predictions that are nearly constant, with values close to the mean of the dataset. This results in a relatively small mean squared error, as the data points are not far from the average.

However, the $R^2$ value is close to zero, indicating that the model explains almost none of the variability in the data. This occurs because the linear model fails to capture any meaningful relationship between the quarter index and GDP growth.

This suggests that, even after removing extreme outliers, the data does not exhibit a strong linear trend. Instead, GDP growth fluctuates around a relatively stable average, making a constant model nearly as effective as a linear one.

However, the model does suggest that GDP growth percentages do tend to cluster around a certain range.


### Method Recommendation for Interpolation

For a policymaker estimating GDP growth between two known data points, cubic spline interpolation is the preferred method over polynomial fitting.

First, in terms of accuracy, cubic splines interpolate the data exactly, ensuring that all known GDP values are preserved. This is particularly important when estimating values between observed quarters, as the interpolant will remain consistent with the data.

Second, cubic splines provide smoothness by enforcing continuity of the function and its first and second derivatives at each knot. This results in a smooth curve without the oscillations often observed in higher-degree polynomial fits.


Finally, cubic splines are less sensitive to outliers than polynomial fits. While splines do pass through all data points, their piecewise structure localizes the effect of extreme values, preventing them from distorting the entire curve. Polynomial fits, however, can be significantly affected by outliers, as a single extreme value can influence the entire model.

Therefore, cubic splines offer a better balance of accuracy, smoothness, and robustness, making them the more appropriate choice for interpolation in this context.

### Big O Analysis and Its Effect on the Recommendation

Big O notation describes how the computational cost of an algorithm grows as the size of the input increases. In this project, the input size is the number of data points \(n\). Big O does not measure the exact runtime, but instead gives the dominant growth rate of the algorithm as the dataset becomes larger.

For the least squares linear model, constructing the design matrix is \(O(n)\), and solving the resulting \(2 \times 2\) system is constant time relative to \(n\). Thus the overall cost is \(O(n)\).

For polynomial fitting of degree \(d\), constructing the Vandermonde matrix requires \(O(nd)\) operations. Forming the normal equations costs approximately \(O(nd^2)\), and solving the resulting \((d+1)\times(d+1)\) linear system costs \(O(d^3)\). Since the degree is small and fixed in this project, such as degree 4, the practical cost is close to linear in \(n\). More generally, however, polynomial fitting becomes more expensive as the degree increases.

For the cubic spline, constructing the tridiagonal system takes \(O(n)\), and solving it with a tridiagonal solver such as Thomas’s algorithm also takes \(O(n)\). Therefore, cubic spline interpolation has overall complexity \(O(n)\).

From a complexity standpoint, both the linear least squares model and the cubic spline are very efficient, and even the degree-4 polynomial fit is manageable for a dataset of this size. Therefore, Big O analysis does not change the recommendation. Cubic spline interpolation is still the best method for estimating GDP growth between two known data points because it is accurate, smooth, and less prone to global oscillation than polynomial fitting.

In other words, although the methods have similar computational feasibility for this project, the recommendation should be based primarily on numerical behavior and modeling quality rather than runtime. Since interpolation is the goal, cubic splines remain the preferred choice.
