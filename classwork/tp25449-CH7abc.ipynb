```python
import numpy as np
import time
```

```python
def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    
    for k in range(max_iterations):
        x_new = x.copy()
        
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
            
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i] 
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
#Challenge 7a
A1 = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)
b1 = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)

x0 = np.zeros_like(b1)
tol = 1e-9  #Raised the tolerance to 10e-9 so it looks cleaner 
max_iterations = 10000


x = np.linalg.solve(A1, b1)
print("Regular Solve: ", x)
```

```python
start_time = time.perf_counter()
solution, iterations = jacobi(A1, b1, x0, tol, max_iterations)
end_time = time.perf_counter()
elapsed_time = end_time - start_time
print(f"Jacobi Solution: {solution}")
print(f"Iterations: {iterations}")
print("Elapsed Time: " , elapsed_time)
```

```python
start_time = time.perf_counter()
solution, iterations = gauss_seidel(A1, b1, x0, tol, max_iterations)
end_time = time.perf_counter()
elapsed_time = end_time - start_time
print(f"Gauss-Seidel Solution: {solution}")
print(f"Iterations: {iterations}")
print("Elapsed Time: " , elapsed_time)
```

```python
#challenge 7d (??? I think???)

def Diag_Dom(A):
    m,n  = A.shape #retrieve dimensions
    
    
    
    for i in range(m):
        diag = abs(A[i,i])
        sum = 0
        for j in range(n): #add up non-diag elements
            if j != i:
                sum += abs(A[i,j])
            #print(diag, sum) this was a sanity check
        if diag < sum: #checks condition for dominance
            return False
        
    return True
                
```
