```python
import random
```

```python
def approximate_pi(numSamples):
    counter = 0
    for i in range(numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1, 1)
        
        if((x**2 + y**2 < 1)): 
            counter += 1
    
    return 4 * counter/numSamples
    
```

```python
print(approximate_pi(100000000))
```

```python
def f1(x,y):
    return (x*x+y*y)

x = np.linspace(-10, 10, 2000)
y = np.linspace(-10, 10, 2000)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
print(Z)
```
