```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def approximate_pi(numSamples):
    counter = 0
    for i in range(numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1, 1)
        
        if((x**2 + y**2 < 1)): 
            counter += 1
    
    return 4 * counter/numSamples
    
```

```python
x = np.arange(0,100)

true_values = []
rel_values = []

for i in range(100):
    test = approximate_pi(1000)
    true_error = abs(np.pi - test)
    rel_error = abs((np.pi - test)/np.pi)
    
    true_values.append(true_error)
    rel_values.append(rel_error)
    
    
true_values = np.array(true_values)
rel_values = np.array(rel_values)
    
plt.plot(x, true_values, "r", label = 'True Error')
plt.plot(x, rel_values, "b", label = 'Relative Error')
plt.legend()
plt.show()
    
```
