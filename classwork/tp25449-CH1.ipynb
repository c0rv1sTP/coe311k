```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(0,10, 1000)
```

```python
plt.subplot(2,2,1)
plt.plot(x, np.exp(x), 'r', label = "Exponential")
plt.subplot(2,2,2)
plt.scatter(x, np.exp(x), label = "Scatter")
plt.subplot(2,2,3)
h = np.random.rand(1000)
plt.hist(h,50, color = "green", label = "Histogram")
plt.subplot(2,2,4)
theta = np.linspace(-1*np.pi,np.pi, 100)
plt.plot(theta, np.arctan(theta), "y-o", label = "Arctan")
plt.legend() # Display the legend.
plt.xlabel('X') # Add a label to the x-axis.
plt.ylabel('Y') # Add a label to the y-axis.
plt.title("Fun Graphs") # Add a graph title.
plt.show()
```
