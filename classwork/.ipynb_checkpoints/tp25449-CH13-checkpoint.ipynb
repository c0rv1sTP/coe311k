```python
import numpy as np
import time
import matplotlib.pyplot as plt

```

```python
def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C
```

```python
sizes = [3 * (2**i) for i in range(8)]
naive_times= []
numpy_times = []

# Store matrix pairs if needed
matrix_pairs = []

for n in sizes:
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)
    matrix_pairs.append((A, B))
    
for A, B in matrix_pairs:
    
    start = time.perf_counter()
    C_naive = naive_matrix_multiplication(A,B)
    end = time.perf_counter()
    naive_times.append(end-start)
    
    start = time.perf_counter()
    C_numpy = A@B
    end = time.perf_counter()
    numpy_times.append(end - start)
    
naive_times= np.array(naive_times)
numpy_times = np.array(numpy_times)

    
print("Naive Times: ", naive_times)
print("Numpy Times: ", numpy_times)

matrix_numbers = list(range(1, 9))

plt.plot(matrix_numbers, naive_times, marker='o', label='Naive')
plt.plot(matrix_numbers, numpy_times, marker='o', label='NumPy Optimized')
plt.xlabel("Matrix Number")
plt.ylabel("Runtime (seconds)")
plt.title("Naive vs NumPy Matrix Multiplication Runtime")
plt.legend()
plt.show()


    
```
