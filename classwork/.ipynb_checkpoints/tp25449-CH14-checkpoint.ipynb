```python
import numpy as np
import matplotlib.pyplot as plt

###This parts is just for reference###

# Define the function and the domain
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi  # Interval
N = 50  # Number of sample points

# Create a uniform grid
x = np.linspace(a, b, N)
y = f(x)

# Plot the sampled points
plt.plot(x, y, 'o-', label='Sampled Points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Uniform Sampling of f(x) = sin(x)')
plt.legend()
plt.show()
```

```python
# Step size
h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
df_dx = (f(x + h) - f(x - h)) / (2 * h)

# Plot the approximation and the true derivative
plt.plot(x, np.cos(x), label='Exact Derivative')
plt.plot(x, df_dx, 'o--', label='Central Difference Approximation')

# Forward Difference Approximation for the derivative
df_dx = (f(x+h) - f(x)) / h
plt.plot(x, df_dx, 'b--', label='Forward Difference Approximation')

# Backward Difference Approximation for the derivative
df_dx = (f(x) - f(x-h)) / h
plt.plot(x, df_dx, 'g--', label='Backward Difference Approximation')




plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Approximating Derivative of f(x) = sin(x)")
plt.legend()
plt.show()
```

Forward difference is a little behind the curve. Backward difference is a little ahead. Central difference sits right on the cosine function.
