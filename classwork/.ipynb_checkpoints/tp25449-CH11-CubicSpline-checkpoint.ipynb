```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Define known data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

# Create a cubic spline interpolation of the data
cs = CubicSpline(x, y, bc_type='natural')

# Generate new x values for a smooth plot
x_interp = np.linspace(0, 5, 100)
y_interp = cs(x_interp)
```

```python
# Plot the original data and cubic spline interpolation
plt.plot(x, y, 'bo', label="Data points")
plt.plot(x_interp, y_interp, 'r-',
label="Cubic Spline Interpolation")
plt.title("Cubic Spline Interpolation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
coefficients = np.polyfit(x, y, 3)
polynomial = np.poly1d(coefficients)
y_pred = polynomial(x_interp)

plt.plot(x, y, 'bo', label="Data points")
plt.plot(x_interp, y_interp, 'r-', label="Cubic Spline Interpolation")

plt.plot(x_interp, y_pred , 'g-', label="Cubic Poly Fit")

plt.title("Cubic Spline vs Cubic Poly Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

Stark difference, poly fit only hits the last 2 points.
