```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def approximate_pi(numSamples):
    counter = 0
    for i in range(numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1, 1)
        
        if((x**2 + y**2 < 1)): 
            counter += 1
    
    return 4 * counter/numSamples
    
```
