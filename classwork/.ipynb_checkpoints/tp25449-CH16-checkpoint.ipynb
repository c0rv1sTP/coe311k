```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Define the parameters

k = 1.0        # Constant in the equation (e.g., spring constant)
h_values = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]      # Step size
x_start = 0    # Start of the interval
x_end = 10     # End of the interval
y_0 = 1.0      # Initial condition for y
v_0 = 0.0      # Initial condition for dy/dx

# Create arrays to store x, y, and v values

def euler_forward(k, h, x_start = 0.0, x_end = 10.0, y_0 = 1.0, v_0 = 0.0):
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

# Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0
    
    
    # Euler's Method loop
    for i in range(1, len(x_values)):
    # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]    

        # Update equations
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)
        
    return x_values, y_values, v_values
    
    
    
plt.figure(figsize=(20, 6))    


for h in h_values:
    x, y, v = euler_forward(k=k, h=h)
    plt.plot(x, y, label=f'h = {h}')

plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.legend()
plt.grid(True)

plt.show()
```

Instability starts at h = 0.1 and greater.

```python
k_values = [0.5, 1, 2, 5, 10]
h = 0.01

plt.figure(figsize=(12, 6))

for k in k_values:
    x, y, v = euler_forward(k=k, h=h)
    plt.plot(x, y, label=f'k = {k}')

plt.xlabel('x')
plt.ylabel('y(x)')
plt.title(f'Forward Euler for different k values with h = {h}')
plt.legend()
plt.grid(True)
plt.show()
```

The periods of the waves become smaller as k grows.

```python
k = 10
h = 0.01

x, y, v = euler_forward(k=k, h=h)

plt.figure(figsize=(12, 6))
plt.plot(x, y, label='Position y(x)')
plt.plot(x, v, '--', label='Velocity v(x)')
plt.xlabel('x')
plt.ylabel('Value')
plt.title('Forward Euler with k = 10, h = 0.01')
plt.legend()
plt.grid(True)
plt.show()
```

Note that the actual solution is oscillatory with CONSTANT amplitude. The increasing amplitude in each of these plots imply that regardless of step size, the solution obtained from Euler's method will eventually become unstable.
