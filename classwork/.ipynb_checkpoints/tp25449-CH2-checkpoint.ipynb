```python
def algo(n):
    bin = 0
    counter = 0    #keeps track of place value
    while(n > 0):
        
        if(n % 2 == 1):
            bin += 10**counter #adds the digit 1 to the left of the current int
            n = n//2
            
        else:
            n = n//2
            
        counter += 1 #moves to next place value
    return bin
```

```python
print(algo(2025))
```
