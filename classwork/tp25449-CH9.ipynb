```python
import numpy as np
import matplotlib.pyplot as plt

# Data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

print("A:\n", A)
print("X: \n" , X)
print("Y: \n", Y)

# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

```python
plt.scatter(X, Y)
x = np.linspace(0, 5, 1000)
plt.title("Least Squares Fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x, m*x+c)
plt.show()
```

```python
from numpy import random

X = np.random.uniform(low=0.0, high=10., size= 10)
X= np.sort(X)

Y= np.random.uniform(low=0.0, high=10., size= 10)
A = np.vstack([X, np.ones(len(X))]).T

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")

plt.scatter(X, Y)
x = np.linspace(0, 10, 1000)
plt.title("Least Squares Fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x, m*x+c)
plt.show()
```

```python
X = np.linspace(0, 1, 11)
noise = np.random.uniform(-0.5, 0.5, len(X)) #Random Pertubation

Y = 2*X + 1.243565 + noise

A = np.vstack([X, np.ones(len(X))]).T

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")

plt.scatter(X, Y)
x = np.linspace(0, 1, 1000)
plt.title("Least Squares Fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x, m*x+c)
plt.show()
```
